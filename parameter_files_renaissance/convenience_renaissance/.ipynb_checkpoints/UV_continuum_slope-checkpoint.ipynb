```python
import matplotlib
matplotlib.use('Agg')

import numpy as np
import matplotlib.pyplot as plt
from hyperion.model import ModelOutput
from astropy.cosmology import Planck13
from astropy import units as u
from astropy import constants
from hyperion.util.constants import pc
import h5py
import os
import sys
import ytree
import yt
np.set_printoptions(threshold=sys.maxsize)
yt.enable_plugins()
from scipy.optimize import curve_fit
```

```python
#get SED
def sed_plot(z, run):
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)


    m = ModelOutput(run)
    wav,flux = m.get_sed(inclination=0,aperture=-1)

    wav  = np.asarray(wav)*u.micron #wav is in micron
    #rest frame = observed / (1+z) => we must be getting rest frame from Powderday so lets convert to observed
    wav *= (1.+z)

    flux = np.asarray(flux)*u.erg/u.s
    dl = Planck13.luminosity_distance(z)
    dl = dl.to(u.cm)

    flux /= (4.*3.14*dl**2.)

#     nu = constants.c.cgs/(wav.to(u.cm))
#     #nu = nu.to(u.Hz)

    flux /= wav

#     #flux density per frequency
#     flux /= nu
#     #convert to a flux density per wavelength
#     flux *= (constants.c.cgs/wav**2)
    #erg / (cm2 Hz s) * (cm/s)
    #erg / (cm Hz micron2 s2)
    # erg / (cm3 s)
    #derg / (cm micron2 s)
    #erg / (Angstrom2 cm s)
    #ergs/s/cm²/Å
    #want erg/ (s nm cm^2)
#     flux = flux.to(u.mJy)
    
    return wav, flux
```

```python
def UV_continuum_slope(wav_func, beta_func, q_func):
    return beta_func * np.log(wav_func) + q_func
```
